```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df1= df.dropna()
```

```python
df1.isnull().sum()
```

```python
df1.Area.unique()
```

```python
df1.columns = ["Region", "Date", "Frequency",	"Estimated Unemployment Rate (%)",
               "Estimated Employed",	"Estimated Labour Participation Rate (%)",	"Area"]
plt.title("Estimated Employed in Rural v/s Urban")
sns.histplot(x="Estimated Employed", hue="Area", data=df1)
plt.show()
```

```python
df1.columns = ["Region", "Date", "Frequency",	"Estimated Unemployment Rate (%)",
               "Estimated Employed",	"Estimated Labour Participation Rate (%)",	"Area"]
plt.title("Unemployment Representation")
sns.histplot(x="Estimated Unemployment Rate (%)", hue="Region", data=df1)
plt.show()
```

```python
df1.columns = ["Region", "Date", "Frequency",	"Estimated Unemployment Rate (%)",
               "Estimated Employed",	"Estimated Labour Participation Rate (%)",	"Area"]
plt.title("Estimated Unemployed in Rural v/s Urban")
sns.histplot(x="Estimated Unemployment Rate (%)", hue="Area", data=df1)
plt.show()
```

```python
unemployment= df1[["Area","Region","Estimated Unemployment Rate (%)"]]
figure=px.sunburst(unemployment, path=["Area","Region"], values="Estimated Unemployment Rate (%)",
                  width=700, height=700,color_continuous_scale="RdY1Gn",
                   title="Unemployment Rate in India(Rural and Urban Oriented)")
figure.show()
```

```python
unemployment= df1[["Region","Area","Estimated Unemployment Rate (%)"]]
figure=px.sunburst(unemployment, path=["Region","Area"], values="Estimated Unemployment Rate (%)",
                  width=700, height=700,color_continuous_scale="RdY1Gn",
                   title="Unemployment Rate in India")
figure.show()
```
